```python

import sys
import os
import numpy as np
#import matplotlib
#matplotlib.use('Agg')  # backend non-interactif pour serveur
import matplotlib.pyplot as plt


T = 1.0
dt = 0.001
M = 5

N = int(T / dt)
paths = np.zeros((M, N + 1))
for i in range(M):
    dW = np.sqrt(dt) * np.random.randn(N)
    paths[i, 1:] = np.cumsum(dW)

time_grid = np.linspace(0, T, N + 1)

results_dir = f"resultat_M={M}"
os.makedirs(results_dir, exist_ok=True)

mean_path = np.mean(paths, axis=0)
variance_path = np.var(paths, axis=0)

# Figure 1: Trajectoires + moyenne
plt.figure(figsize=(10, 6))
for i in range(min(M, 20)):  # limiter le nombre de courbes pour la lisibilité
    plt.plot(time_grid, paths[i], alpha=0.7)
plt.plot(time_grid, mean_path, color='black', linewidth=2, linestyle='--', label='Moyenne Empirique')
plt.title(f"Simulation du Mouvement Brownien (M={M})")
plt.xlabel("Temps")
plt.ylabel("W_t")
plt.grid(True)
plt.legend()
#plt.savefig(os.path.join(results_dir, "brownian_motion_simulation.png"))
plt.show()
plt.close()

# Figure 2: Moyenne
plt.figure(figsize=(10, 6))
plt.plot(time_grid, mean_path, color='black', label='Moyenne Empirique')
plt.title(f"Moyenne Empirique des Trajectoires (M={M})")
plt.xlabel("Temps")
plt.ylabel("Position Moyenne")
plt.grid(True)
plt.legend()
plt.show()
#plt.savefig(os.path.join(results_dir, "mean_path_simulation.png"))
#plt.close()

# Figure 3: Variance vs théorique
plt.figure(figsize=(10, 6))
plt.plot(time_grid, variance_path, label="Variance Empirique")
plt.plot(time_grid, time_grid, label="Variance Théorique (y=t)", linestyle='--')
plt.title(f"Variance Empirique vs Théorique (M={M})")
plt.xlabel("Temps")
plt.ylabel("Variance")
plt.grid(True)
plt.legend()
plt.show()
#plt.savefig(os.path.join(results_dir, "variance_simulation.png"))
#plt.close()

print("Simulation terminée avec succès")
```
